# LABEL ENCODING

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
final['userID_encode'] = le.fit_transform(final["userID"])
final['jobID_encoded'] = le.fit_transform(final["jobID"])
final["tagID_encoded"] = le.fit_transform(final["tagID"])
final["companyID_encoded"] = le.fit_transform(final["companyID"]) 
final["companySize"] = final["companySize"].str.replace("-","")
final["companySize"] = final["companySize"].str.replace("1000 이상","10000000")
final["companySize"] = final["companySize"].fillna(0)

final
```

```python
new_x = final.iloc[:,5:]
new_y = final["applied"]
```

```python
new_x["companySize"].unique()
```

# Step 2:  Apply normalization operation for numerical stability

```python
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()
new_x = standardizer.fit_transform(new_x)
```

# Step 3: Split the dataset into training and testing sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(new_x,new_y,test_size=0.3,random_state=0)
```

# Step 4: Fit a Model to the train data



```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

# Step 5: Make predictions on the testing data

```python
predictions = model.predict(x_test)
```

# Step 6: Calculate the accuracy score by comparing the actual values and predicted values

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predictions)


TN, FP, FN, TP = confusion_matrix(y_test, predictions).ravel()

print('True Positive(TP)  = ', TP)
print('False Positive(FP) = ', FP)
print('True Negative(TN)  = ', TN)
print('False Negative(FN) = ', FN)

accuracy =  (TP+TN) /(TP+FP+TN+FN)

print('Accuracy of the binary classification = {:0.3f}'.format(accuracy))
```
